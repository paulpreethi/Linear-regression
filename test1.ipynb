```python
cldarea=ds.cldarea_cldtyp_mon
cldarea_liq=ds.cldarea_liq_cldtyp_mon
cldarea_ice=ds.cldarea_ice_cldtyp_mon
```

```python
def take_anom(arg):
    anom=(arg.groupby('time.month'))-(arg.groupby('time.month').mean('time'))
    return(anom)
```

```python
cldarea_a=take_anom(cldarea)
cldarea_liq_a=take_anom(cldarea_liq)
cldarea_ice_a=take_anom(cldarea_ice)
```
